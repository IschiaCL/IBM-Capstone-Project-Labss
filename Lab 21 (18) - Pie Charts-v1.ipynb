### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Letâ€™s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# Split the semicolon-separated database lists and count each database
db_series = df['DatabaseWantToWorkWith'].dropna().str.split(';')
db_flat = db_series.explode().str.strip()

# Get the top 5 most-desired databases
top5_db = db_flat.value_counts().head(5)

# Plot the pie chart
plt.figure(figsize=(7,7))
plt.pie(top5_db, labels=top5_db.index, autopct='%1.1f%%', startangle=140, counterclock=False)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')
plt.show()
```

The `DevType` column lists the developer types for respondents. Weâ€™ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
# Split the semicolon-separated developer type lists and count each type
devtype_series = df['DevType'].dropna().str.split(';')
devtype_flat = devtype_series.explode().str.strip()

# Get the top 5 most common developer roles
top5_devtype = devtype_flat.value_counts().head(5)

# Plot the pie chart
plt.figure(figsize=(7,7))
plt.pie(top5_devtype, labels=top5_devtype.index, autopct='%1.1f%%', startangle=140, counterclock=False)
plt.title('Top 5 Developer Roles Among Respondents')
plt.axis('equal')
```

#### Load the data


```python
df = pd.read_csv("survey-data.csv")
```

### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group



- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
# Line Chart of Median ConvertedCompYearly by Age Group

# Group by Age and calculate the median compensation
median_comp_by_age = df.groupby('Age')['ConvertedCompYearly'].median().dropna()

# Plot the line chart
plt.figure(figsize=(10,6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o')
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
# Line chart of median compensation for respondents between ages 25 and 45

# Filter age groups that represent ages 25 to 45
age_labels = [label for label in median_comp_by_age.index if ('25' in label or '30' in label or '35' in label or '40' in label or '45' in label)]

median_comp_25_45 = median_comp_by_age.loc[age_labels]

plt.figure(figsize=(8,5))
plt.plot(median_comp_25_45.index, median_comp_25_45.values, marker='o', color='orange')
plt.title('Median Yearly Compensation (Ages 25-45)')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
# Line Chart of Job Satisfaction (JobSatPoints_6) by Experience Level

# If your dataset has a column like 'WorkExp' or 'YearsCodePro', use it as the experience column.
# Here, we'll use 'WorkExp' if available. Replace with the correct column if needed.

if 'WorkExp' in df.columns:
    exp_col = 'WorkExp'
elif 'YearsCodePro' in df.columns:
    exp_col = 'YearsCodePro'
else:
    exp_col = None

if exp_col:
    # Convert experience column to numeric, coerce errors to NaN
    df[exp_col] = pd.to_numeric(df[exp_col], errors='coerce')
    # Group by experience and calculate median job satisfaction
    exp_sat = df.groupby(exp_col)['JobSatPoints_6'].median().dropna()
    # Plot
    plt.figure(figsize=(10,6))
    plt.plot(exp_sat.index, exp_sat.values, marker='o', color='green')
    plt.title('Median Job Satisfaction (JobSatPoints_6) by Experience Level')
    plt.xlabel('Years of Experience')
    plt.ylabel('Median Job Satisfaction')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("No suitable experience column found in the dataset.")
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here
## Write your code here
# Line Chart of Median ConvertedCompYearly Over Experience Level

# Use 'WorkExp' if available, otherwise try 'YearsCodePro'
if 'WorkExp' in df.columns:
    exp_col = 'WorkExp'
elif 'YearsCodePro' in df.columns:
    exp_col = 'YearsCodePro'
else:
    exp_col = None

if exp_col:
    # Convert experience column to numeric, coerce errors to NaN
    df[exp_col] = pd.to_numeric(df[exp_col], errors='coerce')
    # Group by experience and calculate median compensation
    exp_comp = df.groupby(exp_col)['ConvertedCompYearly'].median().dropna()
    # Plot
    plt.figure(figsize=(10,6))
    plt.plot(exp_comp.index, exp_comp.values, marker='o', color='blue')
    plt.title('Median Compensation (ConvertedCompYearly) by Experience Level')
    plt.xlabel('Years of Experience')
    plt.ylabel('Median Compensation (USD)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("No suitable experience column found in the dataset.")
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here
## Write your code here
# Line Chart of Job Satisfaction (JobSatPoints_6) Across Experience Levels

# Use 'WorkExp' if available, otherwise try 'YearsCodePro'
if 'WorkExp' in df.columns:
    exp_col = 'WorkExp'
elif 'YearsCodePro' in df.columns:
    exp_col = 'YearsCodePro'
else:
    exp_col = None

if exp_col:
    # Convert experience column to numeric, coerce errors to NaN
    df[exp_col] = pd.to_numeric(df[exp_col], errors='coerce')
    # Group by experience and calculate median job satisfaction
    exp_sat = df.groupby(exp_col)['JobSatPoints_6'].median().dropna()
    # Plot
    plt.figure(figsize=(10,6))
    plt.plot(exp_sat.index, exp_sat.values, marker='o', color='purple')
    plt.title('Median Job Satisfaction (JobSatPoints_6) Across Experience Levels')
    plt.xlabel('Years of Experience')
    plt.ylabel('Median Job Satisfaction')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("No suitable experience column found in the dataset.")
```
